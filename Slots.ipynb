```python
import random

print("Welcome to Sam's Slots")
name = input("What is your name?")
print("Hello " + name)
print("Loading account information...")
cashBal = 1000
print("Account balance: ", cashBal)
bet = int(input("How much money do you want to bet?"))
print("You bet " + str(bet))
# play = input("What slot would you like to play 1, 2 3?")
# print("You chose " + play)




def slots(cashBal):
    while cashBal >= 0:
        a = random.randint(1, 9)
        b = random.randint(1, 9)
        c = random.randint(1, 9)
        print(a, b, c)

        if a == b == c:
            print("You won BIG!")
            cashBal += bet ** bet
            print("Your new balance is " + str(cashBal))
            game = input("Play again? yes/no")
            if game == "yes":
                continue
            else:
                break
        elif a == b or a == c or b == c:
            print("You won!")
            cashBal += bet
            print("Your new balance is " + str(cashBal))
            game = input("Play again? yes/no")
            if game == "yes":
                continue
            else:
                break
        else:
            print("You lost!")
            cashBal -= bet
            print("Your new balance is " + str(cashBal))
            game = input("Play again? yes/no")
            if game == "yes":
                continue
            else:
                break
print(slots(cashBal))



# d = input("Would you like to add cash?")
# print("You chose " + d)

```
