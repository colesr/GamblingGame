```python
# A murder mystery game using customer data

import random
import pandas as pd
import sqlite3

conn = sqlite3.connect('[Online_Retail_Data].[dbo]')
df = pd.read_sql_query("select a.*, b.* from [Online_Retail_Data].[dbo].[customers] a inner join [Online_Retail_Data].[dbo].interactions b on  .customer_id = a.customer_id", conn)

WEAPONS = ["Heart", "Cakestand", "Red Bag", "Water Bottle", "Beaker", "Candle"]



def generate_murder():
    """ Defines the murder """

    murder = {
    "date" : [random.date(df.year.min(), df.year.max())],
    "city" : [random.choice(df.city.unique())],
    "country" : [random.choice(df.country.unique())],
    "weapon" : [random.choice(WEAPONS)],
    "suspect" : [random.choice(df.full_name.unique())],
    "victim" : [random.choice(df.full_name.unique())]
    }
    return murder
```
