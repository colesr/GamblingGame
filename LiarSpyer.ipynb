```python
# A murder mystery game using customer data

import random
import pandas as pd
import sqlite3

conn = sqlite3.connect('[Online_Retail_Data].[dbo]')
df = pd.read_sql_query("select a.*, b.* from [Online_Retail_Data].[dbo].[customers] a inner join [Online_Retail_Data].[dbo].interactions b on b.customer_id = a.customer_id", conn)

WEAPONS = ["Heart", "Cakestand", "Red Bag", "Water Bottle", "Beaker", "Candle"]

motives = ["Insurance Fraud", "Embezzlement Cover-up", "Inheritance",
               "Corporate Espionage", "Blackmail", "Insider Trading Secret"]


def generate_murder():
    """ Defines the murder """

    murder = {
    "date" : [random.choice(df.registration_date.max() - df.registration_date.min())],
    "city" : [random.choice(df.city.unique())],
    "country" : [random.choice(df.country.unique())],
    "weapon" : [random.choice(WEAPONS)],
    "suspect" : [random.choice(df.full_name.unique())],
    "victim" : [random.choice(df.full_name.unique())],
    "motive": [random.choice(motives)],
    "case_number": random.randint(1000, 9999)
    }
    return murder

# generate the case

case = generate_murder()
print(case)
```
