```python
# This is a random number generator. Any time that a random number is needed, this simple program
# will produce one.

import random

randNumber = (random.randint(1, 823543)) # generates a random number between 1 and 823543

digit_sum1 = sum(int(digit) for digit in str(randNumber))

digit_sum2 = sum(int(digit) for digit in str(digit_sum1))

MagicNumber = str(digit_sum2)

myNumber = input("What is your favorite number?")
print("Your number choice is " + myNumber)

print("Your MAGIC NUMBER is " + MagicNumber + ".")

if myNumber > MagicNumber:
    print("Yes")
if myNumber == MagicNumber:
    print("You WIN")
elif myNumber < MagicNumber:
    print("No")

```
